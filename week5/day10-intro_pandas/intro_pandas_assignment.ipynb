### Part 2 - Practice with grabbing Data

Let's now get some practice with just grabbing certain parts of data. If you'd like some extra practice, then try answering each of the questions in more than one way (because remember, we can often grab our data in a couple of different ways). 

1. Grab the first 10 rows of the `chlorides` column. 
2. Grab the last 10 rows of the `chlorides` column. 
3. Grab indices 264-282 of the `chlorides` **and** `density` columns. 
4. Grab all rows where the `chlorides` value is less than 0.10. 
5. Now grab all the rows where the `chlorides` value is greater than it's mean (try **not** to use a hard-coded value for the mean, but instead a method).
6. Grab all those rows where the `pH` is greater than 3.0 and less than 3.5. 
7. Further filter the results from 6 to grab only those rows that have a `residual sugar` less than 2.0. 

```python
# Grab the first 10 rows of the chlorides column. 
# These 3 commands work:
wine_df['chlorides'][:10]
wine_df.loc[:10, 'chlorides']
wine_df[:10]['chlorides']
```

```python
# Grab the last 10 rows of the chlorides column. 
# These 2 commands work
wine_df['chlorides'][-10:]
wine_df[-10:]['chlorides']
```

```python
# Grab indices 264-282 of the chlorides and density columns. 
# These 3 commands work:
wine_df[['chlorides', 'density']][264:283]
wine_df[264:283][['chlorides', 'density']]
wine_df.loc[264:283, ['chlorides', 'density']]
```

```python
# Grab all rows where the chlorides value is less than 0.10. 
# These 2 commands work:
wine_df[wine_df['chlorides'] < 0.10]
wine_df.query('chlorides < 0.10')
```

```python
# Now grab all the rows where the chlorides value is greater than it's mean 
# (try not to use a hard-coded value for the mean, but instead a method). 
# These 2 sets of commands work:
wine_df[wine_df['chlorides'] < wine_df['chlorides'].mean()]

chlorides_mean = wine_df['chlorides'].mean()
wine_df.query('chlorides < @chlorides_mean')
```

```python
# Grab all those rows where the pH is greater than 3.0 and less than 3.5. 
# These 2 commands work:
wine_df[(wine_df['pH'] > 3.0) & (wine_df['pH'] < 3.5)]
wine_df.query('pH > 3.0 and pH < 3.5')
```

```python
# Further filter the results from 6 to grab only those rows that have a `residual sugar` 
# less than 2.0. 
# These 2 sets of commands work:
wine_df[(wine_df['pH'] > 3.0) & (wine_df['pH'] < 3.5) & (wine_df['residual sugar'] < 2.0)]


wine_df_renamed = wine_df.rename(columns={"residual sugar": 'residual_sugar'}) # Don't name columns with spaces!
wine_df_renamed.query('pH > 3.0 and pH < 3.5 and residual_sugar < 2.0')
```

### Part 3 - More Practice

Let's move on to just trying to do more complicated things. Use your knowledge of `groupbys`, `sorting`, and the other things that we learned in lecture to answer the following. 

1. Get the average amount of `chlorides` for each `quality` value. 
2. For observations with a `pH` greater than 3.0 and less than 4.0, find the average `alcohol` value by `pH`. 
3. For observations with an `alcohol` value between 9.25 and 9.5, find the highest amount of `residual sugar`. 
4. Create a new column, called `total_acidity`, that is the sum of `fixed acidity` and `volatile acidity`. 
5. Find the average `total_acidity` for each of the `quality` values. 
6. Find the top 5 `density` values. 
7. Find the 10 lowest `sulphates` values. 

```python
# Get the average amount of chlorides for each quality value. 
wine_df.groupby('quality').mean()['chlorides']
```

```python
# For observations with a pH greater than 3.0 and less than 4.0, 
# find the average alcohol value by pH. 
wine_df.query('pH > 3.0 and pH < 4.0').groupby('pH').mean()['alcohol']
```

```python
# For observations with an alcohol value between 9.25 and 9.5, find the highest amount 
# of residual sugar. 
wine_df.query('alcohol >= 9.25 and alcohol <= 9.5')['residual sugar'].max()
```

```python
# Create a new column, called total_acidity, that is the sum of fixed acidity 
# and volatile acidity. 
# These 2 sets of commands work:
wine_df['total_acidity'] = wine_df['fixed acidity'] + wine_df['volatile acidity']

wine_df_renamed = wine_df.rename(columns={'fixed acidity': 'fixed_acidity', 
                                          'volatile acidity': 'volatile_acidity'})
wine_df_renamed.eval('total_acidity = fixed_acidity + volatile_acidity')
```

```python
# Find the average total_acidity for each of the quality values. 
wine_df.groupby('quality').mean()['total_acidity']
```

```python
# Find the top 5 density values. 
wine_df.sort_values('density', ascending=False)['density'][:5]
```

```python
# Find the 10 lowest sulphates values. 
wine_df.sort_values('sulphates')['sulphates'][:10]
```
