# Reverse Polish notation
- https://en.wikipedia.org/wiki/Reverse_Polish_notation

```
class RPN:
  def __init__(self, numbers):
    # set numbers list 
    self.numbers = list(numbers)
      
  def add_number(self, number):
    """
    adds a number to the list of numbers 
    """
    self.numbers.append(number) 
    
  def function(self, funct):
    
    # make sure we have enough numbers to start 
    if len(self.numbers) < 2:
      raise Exception(f'UNABLE TO PREFORM FUNCTION\n{len(self.numbers)} < 2')
    
    # identify the last two numbers from list
    num1 = self.numbers[-1]
    num2 = self.numbers[-2]
    # remove last two numbers from list 
    self.numbers = self.numbers[:-2]
    
    # addition
    if funct == '+':
      new_num = num1 + num2
      self.numbers.append(new_num)
      
    # subtract 
    elif funct == '-':
      new_num = num2 - num1
      self.numbers.append(new_num)
      
    # multiply 
    elif funct == '*':
      new_num = num1 * num2
      self.numbers.append(new_num)

    # divide
    elif funct == '/':
      new_num = num2 / num1
      self.numbers.append(new_num)
      
    else:
      raise Exception('lol, no')
```

```
rpn = RPN(numbers=[1,3,4])

rpn.numbers
```

```
# Function to demonstrate printing pattern triangle 
def triangle(n): 
	
	# number of spaces 
	k = 2*n - 2

	# outer loop to handle number of rows 
	for i in range(0, n): 
	
		# inner loop to handle number spaces 
		# values changing acc. to requirement 
		for j in range(0, k): 
			print(end=" ") 
	
		# decrementing k after each loop 
		k = k - 1
	
		# inner loop to handle number of columns 
		# values changing acc. to outer loop 
		for j in range(0, i+1): 
		
			# printing stars 
			print("* ", end="") 
	
		# ending line after each row 
		print("\r") 

# Driver Code 
n = 5
triangle(n) 
```
